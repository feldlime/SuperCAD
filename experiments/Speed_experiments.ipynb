```python
from sympy import symbols, Eq, lambdify
import sympy
from sympy.mpmath import diff
```

```python
symbols_names = ['symbol_name_1', 'symbol_name_2', 'symbol_name_3']
x, y, z = symbols(' '.join(symbols_names))
equation = Eq(x ** 2 - y*x + x**2 + y ** 2, z ** 0.5 / 2)
syms = [x, y, z]

equation
```

```python
symbols_names = [f'symbol_name_{i}' for i in range(10)]
syms = symbols(' '.join(symbols_names))
equation = Eq(sum([sym ** 2 /  3 for sym in syms]), 25)

equation
```

```python
%%time
eq_new = equation
for _ in range(1000):
    for i, sym_name in enumerate(symbols_names[:-1]):
        eq_new = eq_new.subs(sym_name, i)
print(eq_new)
```

```python
%%time
eq_new = equation
for _ in range(1000):
    eq_new = eq_new.subs([(sym_name, i) for i, sym_name in enumerate(symbols_names[:-1])])
print(eq_new)
```
